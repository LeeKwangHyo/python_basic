```python
import cv2
import matplotlib.pyplot as plt
#matplotlib >>>>> RGB순서
img_basic = cv2.imread("cat.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('cat.jpg')

# 픽셀 수 및 이미지 크기 확인
print(image.shape)
print(image.size)

#이미지 Numpy 객체의 특정 픽셀 가르키기
px = image[100, 100]

#B, G, R 순서로 출력
print(px)

#R값만 출력하기
print(px[2])
```

```python
#특정 범위 픽셀 변경

import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('cat.jpg')

start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]
print("---%s seconds---" %(time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#ROI(관심있는 부분) 추출 및 복사

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')

#Numpy Slicing : ROI 처리 기능
roi = image[0:100, 170:300]

#ROI 단위로 이미지 복사하기
image[0:100, 0:130] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#픽셀별로 색상다루기
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
image[:, :, 2] =0 # 모든 픽셀의 2번 인덱스 = R

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
